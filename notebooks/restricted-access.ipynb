```python
import requests 
import os 
import geopandas as gpd 
```

```python
url = "https://api.canopy.umbra.space/tasking/restricted-access-areas"

headers = {
    "accept": "application/json",
    "authorization": f"Bearer {os.environ['UMBRA_API_TOKEN']}"
}

response = requests.get(url, headers=headers)

gf = gpd.GeoDataFrame.from_features(response.json(), crs='EPSG:4326')
len(gf)
```

```python
gf.explore()
```

```python
gf.to_file("restricted_access_areas.geojson", driver="GeoJSON")
```
